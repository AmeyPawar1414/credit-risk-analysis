```python
import pandas as pd
from sqlalchemy import create_engine
import urllib

# Step 1 - Load CSV
print("Loading CSV...")
df = pd.read_csv(r'C:\Users\namey\credit-risk-analysis\data\raw\application_train.csv')
print(f"✅ CSV loaded — Shape: {df.shape}")

# Step 2 - Connect using faster method
print("Connecting...")
connection_string = (
    "Driver={ODBC Driver 17 for SQL Server};"
    "Server=Amey;"
    "Database=credit_risk_db;"
    "Trusted_Connection=yes;"
    "Encrypt=no;"
)
params = urllib.parse.quote_plus(connection_string)
engine = create_engine(
    f"mssql+pyodbc:///?odbc_connect={params}",
    fast_executemany=True  # ← this is the key, 10x faster
)

with engine.connect() as conn:
    print("✅ Connected!")

# Step 3 - Load with fast_executemany
print("Loading to SQL Server... (should take 1-2 mins now)")
df.to_sql(
    'application_train',
    engine,
    if_exists='replace',
    index=False,
    chunksize=5000  # bigger chunks
)
print("✅ Done! Table created successfully.")
```
